```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=200, n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=20)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y, cmap='winter',s=200)
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,200)
        y_hat = stepFunc(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
def stepFunc(z):
    return 1 if z>0 else 0
```

**Changing value of learning rate shifts line . You may try different learning rate to get better results.**

```python
intercept_,coef_ = perceptron(X,y)
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,200)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(20,10))
plt.plot(x_input,y_input,color='black',linewidth=2)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=200)
plt.ylim(-3,2)
```
